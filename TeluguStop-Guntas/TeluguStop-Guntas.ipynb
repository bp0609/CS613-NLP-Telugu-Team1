```python
# collect all the links from all the pages

all_links = links

while next_button:
    print(len(all_links))
    page = urllib.request.urlopen(next_button["href"]).read()
    soup = BeautifulSoup(page, 'html.parser')
    main_div = soup.find('div', class_='band')
    anchors = main_div.find_all('a', class_ = "read-more", href=True)
    links = [a['href'] for a in anchors]
    all_links.extend(links)
    nav_div = soup.find('div', class_='navigation')
    next_button = nav_div.find('a', class_='next', href=True)
```

```python
# store the links as column of dataframe and save it as csv

import pandas as pd

df = pd.DataFrame(all_links, columns=['links'])
df.to_csv('telugustop_links.csv', index=False)
df
```

```python
def get_links(content):
    soup = BeautifulSoup(content, 'html.parser')
    main_div = soup.find('div', class_='band')
    anchors = main_div.find_all('a', class_ = "read-more", href=True)
    links = [a['href'] for a in anchors]
    return links

def write_to_csv(links, csv_path):
    if not os.path.isfile(csv_path):
        with open(csv_path, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["links"])
    
    with open(csv_path, mode='a', newline='') as file:
        writer = csv.writer(file)
        for link in links:
            writer.writerow([link])

def load_page(url, csv_path):
    page = urllib.request.urlopen(url).read()
    links = get_links(page)
    write_to_csv(links, csv_path)
```

```python
base_url = "https://telugustop.com/telugu/daily-update-categories/telugu-andhra-telangana-tollywood-latest-political-movie-celebrity-live-news-updates-mobile-website/?desktop=yes&amp=yes#google_vignette"
urls = [base_url]

for i in range(1, 1051):
    url = f"https://telugustop.com/telugu/daily-update-categories/telugu-andhra-telangana-tollywood-latest-political-movie-celebrity-live-news-updates-mobile-website/page/{i}?desktop=yes&amp=yes"
    urls.append(url)
```

```python
len(urls)
```

```python
import threading

def main_run(urls, csv_path):
    threads = []
    for url in urls:
        thread = threading.Thread(target=load_page, args=(url, csv_path))
        threads.append(thread)
        thread.start()
    
    for thread in threads:
        thread.join()
```

```python
main_run(urls, 'telugustop_all_links.csv')
```
