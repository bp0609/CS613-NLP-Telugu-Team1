```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
df = pd.read_csv("./TeluguStopLinks-Guntas/TeluguStop-GuntasFinal.csv")
df
```

```python
url = df["links"][1]
```

```python
page = urllib.request.urlopen(url).read()
soup = BeautifulSoup(page, 'html.parser')
```

```python
print(soup)
```

```python
str = ""
for paragraph in soup.find_all('p'):
    str += paragraph.get_text() + "\n"

print(str)
```
